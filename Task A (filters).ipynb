```python
import numpy as np
from matplotlib import pyplot as plt


from skimage.io import imread,imshow,imsave

```

```python

im = imread("C:\\Users\\Abhijeet\\Downloads\\MRI_noisy.tif")
print(im.shape)

plt.figure()
imshow(im)
plt.show()
plt.xticks([]), plt.yticks([])     # Hiding X and Y tick value
```

```python
from scipy.signal import convolve2d

# Mean kernel
ksize = 15
kernel = np.ones((ksize,ksize))
kernel /= kernel.sum()

im_out = convolve2d(im, kernel)

plt.figure()
plt.gray()
plt.subplot(1,2,1)
plt.imshow(im)
plt.subplot(1,2,2)
plt.imshow(im_out)
plt.show()
```

```python
# Median filter
ksize = 5
im_out3 = np.zeros(im.shape)
for y in range(ksize//2, im.shape[0]-ksize//2):
    for x in range(ksize//2, im.shape[1]-ksize//2):
        im_out3[y,x] = np.median(im[y-ksize//2:y+ksize//2+1, x-ksize//2:x+ksize//2+1])

        

plt.figure()
plt.gray()
plt.subplot(1,2,1)
plt.imshow(im)
plt.subplot(1,2,2)
plt.imshow(im_out)
plt.show()
```
