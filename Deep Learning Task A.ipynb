## Deep Learning

```python
##Denoising using deep learning
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.datasets import mnist
```

```python
# load mnist data
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print(X_train.shape)
print(y_train.shape)
```

```python
# print random number from the dataset
index = np.random.randint(0,len(X_train))
plt.imshow(X_train[np.random.randint(index)], cmap = plt.get_cmap('gray'))
```

```python
# reshape train and test data
dim = X_train.shape[1] * X_train.shape[2]
X_train = X_train.reshape(X_train.shape[0], dim).astype('float32')
X_test = X_test.reshape(X_test.shape[0], dim).astype('float32')
```

```python
# normalize the dataset from 0 to 255
X_train = X_train / 255
X_test = X_test / 255
print(X_train.shape)
print(X_test.shape)
```

```python
# introduce noise into train and test data
noise = 0.3
X_train_noisy = X_train + noise * np.random.normal(loc=0.0, scale=1.0, size=X_train.shape)
X_test_noisy = X_test + noise * np.random.normal(loc=0.0, scale=1.0, size=X_test.shape)
X_train_noisy = np.clip(X_train_noisy, 0., 1.)
X_test_noisy = np.clip(X_test_noisy, 0., 1.)
```

```python
# create an autoencode model

denoised_model = Sequential()
denoised_model.add(Dense(500, input_dim= dim, activation='relu'))
denoised_model.add(Dense(300, activation='relu'))
denoised_model.add(Dense(100, activation='relu'))
denoised_model.add(Dense(300, activation='relu'))
denoised_model.add(Dense(500, activation='relu'))
denoised_model.add(Dense(784, activation='sigmoid'))
denoised_model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# train the model
denoised_model.fit(X_train_noisy, X_train, validation_data=(X_test_noisy, X_test), epochs= 100 , batch_size= 150)
```

```python
scores = denoised_model.evaluate(X_test_noisy,y_test)
print("Accuracy: %.2f%%" % (int(scores)))
# accuracy is 27.06
```

```python
# predict on test data
model_pred = denoised_model.predict(X_test_noisy)
```

  ## MSE & SSIM

```python
from skimage.metrics import structural_similarity
import matplotlib.pyplot as plt
import numpy as np
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import mean_squared_error
import cv2
```

```python

def image_comparision(orignal_image, denoised_image):
    mse = mean_squared_error(orignal_image, denoised_image)
    ssim = structural_similarity(orignal_image, denoised_image, multichannel= True) 
    
    fig = plt.figure(figsize = (10,5))
    plt.suptitle("MSE: %.2f, SSIM: %.2f" % (mse, ssim)) # show first image
    ax = fig.add_subplot(1, 2, 1)
    
    plt.imshow(orignal_image, cmap = plt.cm.gray)
    plt.axis("off") # show the second image
    
    ax = fig.add_subplot(1, 2, 2)
    plt.imshow(denoised_image, cmap = plt.cm.gray)
    plt.axis("off") # show the images
    plt.show()
    
    
    
    fig = plt.figure("Images")

    images = ("Original", orignal_image),("denoised", denoised_image)
    
    for (i, (name, image)) in enumerate(images):

        ax = fig.add_subplot(1, 2, i + 1)
        ax.set_title(name)
       # plt.imshow(image, cmap = plt.cm.gray)
        plt.axis("off")# show the figure
        plt.show()# compare the images
```

```python
# print the mse and SSIM of both the original image and the denoised image
image_comparision(orignal_img, blur )
image_comparision(orignal_img_med, median)
image_comparision(orignal_img_wave, denoised_image_wave)
```
