```python
import numpy as np
from matplotlib import pyplot as plt
from skimage.io import imread,imshow,imsave
from matplotlib import pyplot as plt
%matplotlib notebook
import cv2
from skimage.util import random_noise
```

```python
orignal_img = imread("C:\\Users\\Abhijeet\\Downloads\\task2.tif")    # Loading Image
print(orignal_img.shape)

plt.figure()
imshow(orignal_img)
plt.show()
plt.xticks([]), plt.yticks([])     # Hiding X and Y tick value
```

```python
# Adding Noise
noisy_img = random_noise(orignal_img, seed=None, clip=True)
plt.figure()
imshow(noisy_img)
plt.show()
```

```python
#Mean Filter
from scipy.signal import convolve2d

# Mean kernel
ksize = 15
kernel = np.ones((ksize,ksize))
kernel /= kernel.sum()

denoised_img = convolve2d(noisy_img, kernel)

plt.figure()
plt.gray()
plt.subplot(1,2,1)
plt.imshow(noisy_img[::5, ::5])
plt.subplot(1,2,2)
plt.imshow(denoised_img[::5,::5])
plt.show()
```
